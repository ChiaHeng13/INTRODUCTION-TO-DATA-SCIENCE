#MLB球員表現及其他因素與球隊勝場之關係

###一、研究動機

棒球，做為國球往往在國際賽中在國內引起熱烈討論，不論是2013年世界棒球經典賽、2014年21U世界棒球賽……一場又一場的比賽使球迷留下深刻回憶。棒球是項團隊運動，個人的表現未必能明顯地影響球隊，尤其在賽季相當長的職棒比賽中，而本組希望透過數據分析，探討團隊中球員表現，和各項因素是否能顯著影響球隊勝場，而又以美國職棒大聯盟(MLB)各球隊為研究目標，透過此研究希望能預期未來球隊在單一球季勝場數，以供各界參考用。

###二、研究目的

*   1.探討團隊薪資高低對於球隊例行賽勝場數之影響。
* 2.探討各項數據與球隊賽季表現的影響
*  3.透過此研究實證結果，提供球迷對球隊預期勝場數之推算。



###三、文獻回顧

許元俊(2016)利用美國職棒大聯盟(MLB)近20 個球季 (1996至2015年)的球隊與球員資料作為樣本，探討球隊薪資結構，顯示出前一年勝率與平均年齡與球隊勝率間存在著顯著影響，研究中吉尼係數每增加1單位會使球隊勝率下降，對於球隊勝率存在著負向影響。

從固定效果中顯著值中，同區球隊的效果較為顯著，也有同區球隊的效果較不顯著，本組認為這與聯盟制度有關，由於聯盟分為國家聯盟與美國聯盟，各聯盟再細分三區，由於季後賽是由各區第一與外卡冠軍所進行，且同區球隊在賽程安排上較容易相遇，因此本組認為同區球隊間的影響性較大，彼此不同區的交互影響程度較小。

###七、結論與政策建議

棒球球員個人能力在球季期間影響程度相對其他運動小，即使個人能力突出，團隊仍和是關鍵，所以安打率和全壘打等打擊數據對勝場影響較不顯著，反觀得失分較為顯著也代表是影響勝利的主要關鍵之一。在球團經營方面，雖然主場票房顯著值相對不顯著，但接近臨界值，進場人數多幫助球員加油，能增加球員在主場勝率，有正面影響效果。然而，該研究仍然存在著遺漏變數，在棒球這項運動中逐漸發展出運動科學，但有部分影響為數據無法計算的，例如球員在跨主場移動中所產生的疲勞，亦或者是球員在守備上隱形的瑕疵……這些無法用數值衡量的變數皆影響比賽結果。
藉由該份研究，本組期望提供球迷一個推算球隊勝利的指標，對於職棒球隊而言，沒有獲得總冠軍即為一個失敗的球季，能否進入晉後賽勝場是個重要因素，希望能利用此研究得出總勝場的方程式，供球迷參考。此外，也可提供給球團，在衡量球隊未來經營方向，是否放眼未來或是全力衝刺，將今年勝場預估和其他因素做出最適決策。
